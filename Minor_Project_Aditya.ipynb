#CODE

```python

import csv

def add_expense():
  amount = float(input("Enter the amount: "))
  category = input("Enter the category: ")
  date = input("Enter the date (YYYY-MM-DD): ")
  expense = {"amount": amount, "category": category, "date": date}
  expenses.append(expense)
  print("Expense added successfully!")

def view_summary():
  print("Summary:")
  for category in categories:
    total = sum([expense["amount"] for expense in expenses if expense["category"] == category])
    print(f"{category}: {total}")
  total_spending = sum([expense["amount"] for expense in expenses])
  print(f"Total Spending: {total_spending}")

def save_data():
  with open("expenses.csv", "w", newline="") as csvfile:
    fieldnames = ["amount", "category", "date"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(expenses)
  print("Data saved to expenses.csv")

def load_data():
  global expenses
  try:
    with open("expenses.csv", "r") as csvfile:
      reader = csv.DictReader(csvfile)
      expenses = [row for row in reader]
      print("Previous expenses loaded.")
  except FileNotFoundError:
    print("No previous expenses found.")

def main():
  global expenses, categories
  expenses = []
  categories = set()
  load_data()

  while True:
    print("1. Add expense")
    print("2. View summary")
    print("3. Save data")
    print("4. Exit")
    choice = int(input("Enter your choice: "))

    if choice == 1:
      add_expense()
    elif choice == 2:
      view_summary()
    elif choice == 3:
      save_data()
    elif choice == 4:
      break
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
  main()
```
